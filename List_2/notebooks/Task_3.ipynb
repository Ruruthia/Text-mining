```python
from itertools import product

import editdistance
import morfeusz2

from utils import score_documents, prepare_question
```

```python
with open("../outputs/indexed_contents.txt", "r") as f:
    content_lines = f.readlines()
with open("../outputs/indexed_titles.txt", "r") as f:
    titles_lines = f.readlines()
```

```python
morph = morfeusz2.Morfeusz()
```

```python
def scaled_editdist(ans, cor):
    ans = ans.lower()
    cor = cor.lower()

    return editdistance.eval(ans, cor) / len(cor)
```

```python
def answer(question, weights):
    _, question = prepare_question(question, morph)
    question = [token for token in question if len(token) > 1]
    while question:
        query = ' '.join(q for q in question)
        search_results = list(score_documents(query, weights, morph).keys())[:7]
        print([titles_lines[result] for result in search_results])
        for result in search_results:
            title = titles_lines[result]
            _, res_tokens = prepare_question(title, morph)
            for t1, t2 in product(res_tokens, question):
                if scaled_editdist(t1, t2) <= 0.3:
                    break
            else:
                paren_index = title.find('(')
                if paren_index != -1:
                    title = title[:paren_index]
                return title

        # if answer not found, remove first token of query
        del question[0]
    return 'nie mam pojęcia, sorry \n'
```

```python
with open('../data/pytania.txt', 'r') as f_in:
    with open('../data/moje_odpowiedzi.txt', 'a') as f_out:
        line = f_in.readline()
        while line:
            a = answer(line.strip(), [-0.5, 2, -0.5, 0.5, 2.5])
            print(a.strip()+'/n')
            f_out.write(a)
            line = f_in.readline()
```
