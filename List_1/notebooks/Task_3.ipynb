```python
import html
import unicodedata
```

```python
import nltk
nltk.download('punkt')
from nltk.tokenize import word_tokenize
from transformers import *
```

```python
with open("../data/cytaty.txt", "r") as f:
    t0 = f.read()
```

```python
def split_by_space(text):
    start = 0
    stop = 1
    splitted = []
    while stop < len(text):
        if unicodedata.category(text[stop]) == 'Zs':
            if stop > start:
                splitted.append(text[start:stop])
            start = stop + 1
        stop += 1
    return splitted

def preprocess(text):
    return [html.unescape(word.strip()) for word in text]

def strip_punctuation(text):
    stripped = []
    for word in text:
        if word:
            start = 0
            while start < len(word) and unicodedata.category(word[start])[0] == 'P' :
                start += 1
            stop = len(word) - 1
            while stop > 1 and unicodedata.category(word[stop])[0] == 'P':
                stop -= 1
            if stop - start > 0:
                stripped.append(word[start:(stop+1)])
    return stripped

def postprocess(text):
    return [word.strip() for word in text]
```

```python
t1 = split_by_space(t0)
t2 = preprocess(t1)
t3 = postprocess(strip_punctuation(t2))
```

```python
t4 = word_tokenize(t0[:1000000], language='polish')
```
